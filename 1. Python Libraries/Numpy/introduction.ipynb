# Introduction to NumPy

### What is NumPy?

- A Python library for numerical and mathematical operations

- Provides support for multi-dimensional arrays

- Much faster than Python lists (written in C)


### Why NumPy?

- Efficient memory usage

- Vectorized operations (no explicit loops)

- Foundation for libraries like Pandas, SciPy, TensorFlow
