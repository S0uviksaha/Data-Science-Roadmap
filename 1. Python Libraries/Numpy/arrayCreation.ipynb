# ***Array Creation methods***

### *Creating NumPy Arrays*

- Arrays can be created using 
    - array(), 
    - zeros(), 
    - ones(), 
    - arange(), 
    - linspace().

```python
import numpy as np
```

```python
# using array()
arr = np.array([1, 2, 3, 4, 5])
arr
```

```python
# using zeros()
arr1 = np.zeros(2, dtype=int)
arr1
```

```python
arr2 = np.zeros((2,3), dtype=int)
arr2
```

```python
# using ones()
arr3 = np.ones((3,4), dtype=int)
arr3
```

```python
# using arange()
arr4 = np.arange(5,15,2)
arr4
```

```python
# using linspace()
arr5 = np.linspace(0,21,6)          # evenly spaced values
arr5
```
